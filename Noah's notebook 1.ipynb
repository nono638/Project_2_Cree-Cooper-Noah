```python
import pandas as pd
from sqlalchemy import create_engine
imdb_movies_str = "IMDb movies.csv"
imdb_movies_df = pd.read_csv(imdb_movies_str)
imdb_movies_df.columns
```

```python
imdb_movies_df2 = imdb_movies_df[["title","year","genre","country","avg_vote",\
                                ]].copy()
imdb_movies_df2
```

```python
imdb_movies_df2_USA = imdb_movies_df2.loc[imdb_movies_df2["country"]=="USA"]
imdb_movies_df2_USA
```

```python
imdb_movies_df2_USA_Post1970 = imdb_movies_df2_USA.loc[imdb_movies_df2_USA["year"]>=1970]


imdb_movies_df2_USA_Post1970["year"] =imdb_movies_df2_USA_Post1970["year"].astype(str)
imdb_movies_df2_USA_Post1970['YearTitle'] = imdb_movies_df2_USA_Post1970[['year', 'title']].apply(lambda x: '_'.join(x), axis=1)
imdb_movies_df2_USA_Post1970=imdb_movies_df2_USA_Post1970[["YearTitle","title",\
                                                          "year","genre",\
                                                          "country","avg_vote"]]

#imdb_movies_df2_USA_Post1970=imdb_movies_df2_USA_Post1970.set_index("YearTitle")



imdb_movies_df2_USA_Post1970=imdb_movies_df2_USA_Post1970.rename(columns={"YearTitle":"yeartitle",\
                                                                         "year":"yearog"})

imdb_movies_df2_USA_Post1970.drop_duplicates("yeartitle", keep="first", inplace=True)

imdb_movies_df2_USA_Post1970.head()
#This is the final IMDB df
```

```python
rotten_string = "rotten_tomatoes_movies.csv"
rotten_df = pd.read_csv(rotten_string)
rotten_df.columns
```

```python
rotten_df2 = rotten_df[["movie_title","genre","audience_rating",\
                        "tomatometer_rating","in_theaters_date"]]
#rotten_df3.loc[]
rotten_df2.rename(columns = {"movie_title":"title","in_theaters_date":"year"}, inplace=True)
#rotten_df2.set_index("title", inplace=True)
rotten_df2
```

```python
rotten_with_valid_years = rotten_df2[rotten_df2["year"].notnull()]

rotten_with_valid_years["year"] =rotten_with_valid_years["year"].astype(int)

rotten_with_valid_years = rotten_with_valid_years.loc[rotten_with_valid_years["year"]>=1970]

rotten_with_valid_years["year"] =rotten_with_valid_years["year"].astype(str)
rotten_with_valid_years
```

```python

rotten_with_valid_years['yeartitle'] = rotten_with_valid_years[['year', 'title']].apply(lambda x: '_'.join(x), axis=1)

rotten_with_valid_years
```
